##### CÁLCULO DA EQUAÇÃO DIFERENCIAL USANDO O MÉTODO DE HEUN

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

##### entrada de dados

```python
def f(z,x):
    return (x/z)*(1-x/z)
def g(x):
    return x/(1+np.log(x))
h = 0.2
a = 1
b = 2
t = np.linspace(a,b,int(np.round((b-a)/h+1)))

s0 = 1 
```

##### método de heun

```python
s = np.zeros(len(t))
s[0] = s0
print("i : %3d, t : %5.4f, w : %5.11f, y : %5.11f" % (1, t[0],s[0],g(t[0])))

for i in range(0, len(t)-1):
    s[i + 1] = s[i] + h/2*(f(t[i], s[i])+f(t[i+1],s[i]+h*f(t[i], s[i])))
    print("i : %3d, t : %5.4f, w : %5.11f, y : %5.11f" % (i+2, t[i+1],s[i+1],g(t[i+1])))
```

##### representação gráfica

```python
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, g(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
#plt.ylabel('f(t,s)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
