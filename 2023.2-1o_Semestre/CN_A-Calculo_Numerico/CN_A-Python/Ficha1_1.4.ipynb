```python
import numpy as np
import matplotlib.pyplot as plt
import sympy

x=np.arange(8)
y=(26-4*x)/13
z=(20-3*x)/10

plt.plot(x, y, z)
plt.title("Plot line in Matplotlib",fontsize=15)
plt.xlabel("X",fontsize=13)
plt.ylabel("Y",fontsize=13)
plt.show()
```

```python
x=np.arange(8)
y=(25.9-4*x)/13
z=(20.1-3*x)/10

plt.plot(x, y, z)
plt.title("Plot line in Matplotlib",fontsize=15)
plt.xlabel("X",fontsize=13)
plt.ylabel("Y",fontsize=13)
plt.show()
```

```python
a, b = sympy.symbols('a b')
sympy.solve((4*a+13*b-26, 3*a+10*b-20),(a, b))
```

```python
sympy.solve((4*a+13*b-25.9, 3*a+10*b-20.1),(a, b))
```

```python
a = np.array([[4, 13], [3, 10]])
b = np.array([26, 20])
x = np.linalg.solve(a, b)
x
```
