### CÁLCULO DA EQUAÇÃO DIFERENCIAL USANDO O MÉTODO DE EULER

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

##### entrada de dados

```python
def f(z,x):
    return 1-x/z
def g(x):
    return (4+x**2)/(2*x)
h = 0.1
a = 2
b = 3
t=np.linspace(a,b,int((b-a)/h+1))

s0 = 2  
```

##### método de euler

```python
s = np.zeros(len(t))
s[0] = s0

print("i : %3d, t : %5.4f, w : %5.11f, y : %5.11f" % (1, t[0],s[0],g(t[0])))

for i in range(0, len(t)-1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
    print("i : %3d, t : %5.4f, w : %5.11f, y : %5.11f" % (i+2, t[i+1],s[i+1],g(t[i+1])))
```

##### representação gráfica 

```python
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, g(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
